```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image as img
```

```python
def right_shift(fname, k):
    image = img.imread(fname=fname)
    result = image.copy() 
    if(k == 0):
      return result
    for h in range(0, image.shape[0]):
        for w in range(0, image.shape[1]):
            for l in range(0, image.shape[2]):
                result[h][w][l] = min([image[h][w][l] + k, 255])
    return result
```

```python
def plot_images(images, filename) : 
  fig, axs = plt.subplots(1, 4, figsize=(32,18)) 
  for i, ax in enumerate(axs) : 
      ax.imshow(images[i]) 
      ax.set_title(f"{filename}, k = {K[i]}", fontsize=22) 
      ax.axis('off') 
  plt.show()
```

```python
files = ['low_contrast.jpg', 'high_contrast.jpeg', 'low_detalization.jpg', 'high_detalization.jpg']
K = [0, 25, 50, 100]

images = []
for file in files:
  for k in K:
    images.append(right_shift(file, k))
  plot_images(images, file)
  images.clear()
```
